## SVD Compression

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
# plt.style.use('seaborn-poster')


from matplotlib import rcParams
from scipy import integrate
from mpl_toolkits.mplot3d import Axes3D


from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten, MaxPool2D
from keras import optimizers
from keras.layers import Activation
from keras.utils.generic_utils import get_custom_objects
from keras import backend as K


rcParams.update({'font.size': 16})
plt.rcParams['figure.figsize'] = [8, 8]
sns.set_context("paper",font_scale = 1.4) 
```

### 1. Read the image 

```python
image = cv.imread('image.jpg')
plt.axis('off')
plt.title('Yves with Avatar face :-)')
plt.imshow(image);
```

### 2. Convert the colored image into a gray level image

```python
GrayImg = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
plt.axis('off')
plt.title('The gray level image')
plt.imshow(GrayImg,cmap='gray');
```

###  3. Show how many unique elements does image matrix from (ii) have

```python
np.unique(GrayImg).size
```

### 4.  Perform the SVD on the image-matrix of the gray-level image

```python
U,s,V = np.linalg.svd(GrayImg, full_matrices=False)

S = np.diag(s) # We create the diagonal matrix of the singular values
```

### 5. Approximate gray level image matrix by its SVD

```python
pos=1
for k in (1,3,5):
    r=int(k*len(s)/100) #We find the percentage of singular values to be considered 
    ApproxImage = U[:,:r]@S[0:r,0:r]@V[:r,:]
    plt.subplot(1,3,pos)
    plt.title('{} % displayed'.format(k))
    plt.imshow(ApproxImage,cmap='gray')
    plt.axis('off')
    pos=pos+1
```
